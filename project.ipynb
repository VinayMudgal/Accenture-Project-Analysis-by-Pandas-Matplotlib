# ACCENTURE SOCIAL BUZZ DATA ANALYSIS 

### Import all Necessery libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

#### Load Content file DataSet

```python
Content_df= pd.read_csv("Content.csv")
Content_df
```

```python
Content_df.info()
```

```python
Content_df.describe()
```

### Data Cleaning - Content file DataSet

##### Remove unnecessery Columns & Clean the Data

```python
Content_df = Content_df.drop(columns=["URL", "Unnamed: 0"])

Content_df
```

```python
Content_df["Category"] = Content_df["Category"].str.replace('"','')
Content_df["Category"] = Content_df["Category"].str.lower()

Content_df
```

##### Check and Remove Null Rows

```python
Content_df = Content_df.dropna()

Content_df
```

```python
Content_df["Content Type"] = Content_df["Type"]

Content_df = Content_df.drop(columns="Type")

Content_df
```

#### Load Reaction file

```python
reaction_df= pd.read_csv("Reactions.csv")
reaction_df
```

```python
reaction_df.info()
```

```python
reaction_df.describe()
```

#### Data Cleaning for Reaction File

```python
reaction_df.drop(columns="Unnamed: 0")
reaction_df= reaction_df.drop_duplicates()

reaction_df
```

```python
reaction_df["nullreaction"] = reaction_df["Type"].isnull()

reaction_df["usernull"]= reaction_df["User ID"].isnull()

reaction_df
```

```python
reaction_df = reaction_df.loc[(reaction_df["nullreaction"]==False) & (reaction_df["usernull"] == False)]

reaction_df
```

##### Remove unnecessery Columns

```python
reaction_df = reaction_df.drop(columns=["nullreaction","usernull","Unnamed: 0"])

reaction_df
```

```python
reaction_df["Reaction Type"]= reaction_df["Type"]

reaction_df = reaction_df.drop(columns="Type")

reaction_df
```

#### Load Reaction Type file DataSet

```python
reactiontype_df = pd.read_csv("ReactionTypes.csv")
reactiontype_df
```

```python
reactiontype_df["Reaction Type"] = reactiontype_df["Type"]

reactiontype_df = reactiontype_df.drop(columns= ["Type" , 'Unnamed: 0'])

reactiontype_df
```

### Joining all DataSets 

```python
join_df = Content_df.merge(reaction_df,on="Content ID", how="inner")

join_df
```

```python
df  = join_df.merge(reactiontype_df, on ="Reaction Type", how="inner")

df
```

#### Getting the final DataFrame

```python
df
```

#### Remove unnecessery columns from Final DataFrame

```python
df= df.drop(columns=["User ID_x", "User ID_y"])

df
```

##### Convert Final data Frame in to CSV file 


```python
df.to_csv("finalData.csv", index=False)
```

#### Analysis for the Top 5 categories according to the score

```python
df
```

```python
edf = df.groupby("Category").agg(
    High_score=("Score", sum)
)

edf = edf.reset_index()

edf = edf.sort_values(by = "High_score", ascending=False)

edf = edf.head(5)
edf
```

#### Export the File `Top 5 Categories`

```python
edf.to_csv("Top5categories.csv", index=False)
```

#### Finding day for the Highest Content 

```python
df
```

```python
kdf = df.groupby(['Day' , 'Content Type', 'Category']).agg(
    All_Content = ('Content Type' , 'count')
    
)

kdf = kdf.reset_index()

kdf = kdf.sort_values('All_Content' , ascending=False)
kdf
```

```python
sns.catplot(x = 'Day' , y = 'All_Content' , data = kdf , hue = 'Content Type')


plt.grid()
plt.title('Total Contents Vs Days')
plt.xticks(rotation = 35)
plt.show()
```
